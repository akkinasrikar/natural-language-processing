```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
#model predicts close to 1 but actual prediction is 0 --> cost is large
-1*(0*np.log(0.9999)+(1-0)*np.log(1-0.9999))
```

```python
#model predicts close to 1 and actual prediction is 1
-1*(1*np.log(0.9999)+(1-1)*np.log(1-0.9999))
```

```python
#model predicts close to 0 but actual prediction is 1
-1*(1*np.log(0.0001)+(1-0)*np.log(1-0.0001))
```

```python
#model predicts close to 0 and actual prediction is 0
-1*(0*np.log(0.0001)+(1-0)*np.log(1-0.0001))
```

```python
def GradientDescent(x,y,theta,alpha,epochs):
    m = x.shape[0] #no of rows
    cost=[]
    for i in range(epochs):
        z = np.dot(x,theta)
        h = sigmoid(z)
        J = -1/m * (np.dot(y.transpose(),np.log(h)) + np.dot((1-y).transpose(),np.log(1-h)))
        print(f"Epoch : {i+1} Cost : {J[0][0]}")
        cost.append(J[0][0])
        theta = theta - (alpha/m)*np.dot(x.transpose(),(h-y))
    return float(J),theta,cost
```

```python
def extract_features(tweet,freqs):
    words = process_tweets(tweet)
    x =  np.zeros((1,3))
    x[0][0] = 1
    for word in words:
        x[0,1] += freqs.get((word, 1.0),0)
        x[0,2] += freqs.get((word, 0.0),0)
    return x    
```

```python
epochs=1000
learning_rate=1e-7
J,theta,cost = GradientDescent(X,Y,np.zeros((3,1)),learning_rate,epochs)
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(epochs),cost)
plt.xlabel("Epochs")
plt.ylabel("Cost Error")
```

```python
def predict_tweet(tweet,freqs,theta):
    x = extract_features(tweet,freqs)
    y_pred = sigmoid(np.dot(x,theta))
    return y_pred
```

```python
def test_logistic_regression(test_x,test_y,freqs,theta):
    y_preds=[]
    for i in test_x:
        y_pred=predict_tweet(i,freqs,theta)
        if y_pred>=0.5:
            y_preds.append(1)
        else:
            y_preds.append(0)
    accuracy = (y_preds==np.squeeze(test_y)).sum()/len(test_x)
    return accuracy
```

```python
acc=test_logistic_regression(test_x,test_y,freqs,theta)
```

```python
print(f"Accuracy is {acc}")
```
